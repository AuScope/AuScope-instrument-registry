```python
import json
from pathlib import Path

ROR_JSON = Path(r"C:\Drive_D\projects\avre\apps\PIDINST\AuScope-instrument-registry\ckan_batch\kill\v2.3-2026-02-24-ror-data.json")

SCIENTIFIC_TYPES = {
    "facility",
    "education",
    # "government",
    # "nonprofit",
    # "archive",
    # "healthcare",
}

with ROR_JSON.open("r", encoding="utf-8") as f:
    data = json.load(f)

# depending on the dump file, records may be the top-level list
# or under a key like "items"
records = data["items"] if isinstance(data, dict) and "items" in data else data

au_scientific = []

for rec in records:
    locations = rec.get("locations", [])
    is_au = any(
        loc.get("geonames_details", {}).get("country_code") == "AU"
        for loc in locations
    )
    if not is_au:
        continue

    types = set(rec.get("types", []))
    if not (types & SCIENTIFIC_TYPES):
        continue

    display_name = next(
        (
            n["value"]
            for n in rec.get("names", [])
            if "ror_display" in n.get("types", [])
        ),
        None,
    )

    website = next(
        (l["value"] for l in rec.get("links", []) if l.get("type") == "website"),
        None,
    )

    state = next(
        (
            loc.get("geonames_details", {}).get("country_subdivision_name")
            for loc in locations
            if loc.get("geonames_details", {}).get("country_code") == "AU"
        ),
        None,
    )

    au_scientific.append(
        {
            "ror_id": rec.get("id"),
            "name": display_name,
            "types": sorted(types),
            "website": website,
            "state": state,
        }
    )

print(f"Found {len(au_scientific)} Australian scientific/facility records")
```
